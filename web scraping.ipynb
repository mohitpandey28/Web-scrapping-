```python
import pandas
import requests
from bs4 import BeautifulSoup
```

```python
name=[]
price=[]
rating=[]

for i in range(1,20):
  print(i)
  url='https://www.flipkart.com/mobiles/mi~brand/pr?sid=tyy%2C4io&otracker=nmenu_sub_Electronics_0_Mi&page='+str(i)

  req= requests.get(url)
  req.text

  soup=BeautifulSoup(req.text,'html.parser')
  box = soup.find("div",attrs={"class":"_1YokD2 _3Mn1Gg"})

  product_name = box.find_all("div",attrs={"class":"_4rR01T"})
  for i in product_name:
     product_name=i.text
     name.append(product_name)

  product_price = box.find_all("div",attrs={"class":"_30jeq3 _1_WHN1"})
  for i in product_price:
    product_price=i.text
    price.append(product_price)

  product_rating = box.find_all("div",attrs={"class":"_3LWZlK"})
  for i in product_rating:
    product_rating=i.text
    rating.append(product_rating)

  import pandas as pd
  df=pd.DataFrame({'product_name':name,'price':price,'rating':rating})
  print(df)
```
